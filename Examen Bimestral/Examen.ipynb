```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider

def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3

def Spline_deriv(x: float, x0: float, pars: dict[str, float]) -> float:
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return b + 2 * c * (x - x0) + 3 * d * (x - x0) ** 2

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    ax.plot(_x, _y, color="red")

# Inicializar la posición de la tangente
x_tangent = 0
y_tangent = 5
slope = Spline_deriv(x_tangent, xs[1], s[0])

# Definir la línea tangente con un rango muy corto
tangent_x = np.linspace(x_tangent - 0.1, x_tangent + 0.1, 10)
tangent_y = y_tangent + slope * (tangent_x - x_tangent)
tangent_line, = ax.plot(tangent_x, tangent_y, color="blue", linestyle="--")

ax.scatter(xs, ys)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Interpolación con splines cúbicos y línea tangente")

# Crear el slider para rotar la tangente
ax_slider = plt.axes([0.1, 0.1, 0.8, 0.05], facecolor='lightgoldenrodyellow')
slider = Slider(ax_slider, 'Ángulo', -90.0, 90.0, valinit=np.degrees(np.arctan(slope)))

# Función para actualizar la línea tangente
def update(val):
    angle = np.radians(slider.val)
    slope = np.tan(angle)
    tangent_x = np.linspace(x_tangent - 0.1, x_tangent + 0.1, 10)
    tangent_y = y_tangent + slope * (tangent_x - x_tangent)
    tangent_line.set_xdata(tangent_x)
    tangent_line.set_ydata(tangent_y)
    fig.canvas.draw_idle()

slider.on_changed(update)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider

def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3

def Spline_deriv(x: float, x0: float, pars: dict[str, float]) -> float:
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return b + 2 * c * (x - x0) + 3 * d * (x - x0) ** 2

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    ax.plot(_x, _y, color="red")

# Inicializar la posición de la tangente
x_tangent = 0
y_tangent = 5
target_y = 4
slope = (target_y - y_tangent) / (0.1)  # Calcular la pendiente para apuntar hacia y = 4

# Definir la línea tangente con un rango muy corto
tangent_x = np.linspace(x_tangent - 0.1, x_tangent + 0.1, 10)
tangent_y = y_tangent + slope * (tangent_x - x_tangent)
tangent_line, = ax.plot(tangent_x, tangent_y, color="blue", linestyle="--")

ax.scatter(xs, ys)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Interpolación con splines cúbicos y línea tangente")

# Crear el slider para ajustar la pendiente de la tangente
ax_slider = plt.axes([0.1, 0.1, 0.8, 0.05], facecolor='lightgoldenrodyellow')
slider = Slider(ax_slider, 'Pendiente', -10.0, 10.0, valinit=slope)

# Función para actualizar la línea tangente
def update(val):
    slope = slider.val
    tangent_y = y_tangent + slope * (tangent_x - x_tangent)
    tangent_line.set_ydata(tangent_y)
    fig.canvas.draw_idle()

slider.on_changed(update)

plt.show()
```
