# SPINES CUBICOS. -

### EJEMPLO DEL CODIGO. -

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("InterpolaciÃ³n con splines cÃºbicos")
plt.show()
```

### CREACION DE LAS GRAFICAS PARA EL EXAMEN. -

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Definir los puntos de control
puntos_x = np.array([-1, 0, 1])
puntos_y = np.array([1, 5, 3])

# Crear el spline cÃºbico
spline = CubicSpline(puntos_x, puntos_y)

# Crear un rango de valores x para graficar el spline
x_vals = np.linspace(-1.5, 1.5, 100)
y_vals = spline(x_vals)

# Calcular la pendiente en el punto (0, 5)
x_tangente = 0
y_tangente = spline(x_tangente)
pendiente_tangente = spline.derivative()(x_tangente)

# Definir la lÃ­nea tangente
x_tangent_vals = np.array([-0.5, 0.5])  # Un rango pequeÃ±o alrededor de x=0
y_tangent_vals = y_tangente + pendiente_tangente * (x_tangent_vals - x_tangente)

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='Spline CÃºbico', color='blue')
plt.scatter(puntos_x, puntos_y, color='red', label='Puntos de Control')
plt.plot(x_tangent_vals, y_tangent_vals, linestyle='--', color='green', label='LÃ­nea Tangente en (0, 5)')
plt.title('GrÃ¡fica del Spline CÃºbico con LÃ­nea Tangente')
plt.xlabel('x')
plt.ylabel('S(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```
